```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

# Load your dataset
df = pd.read_csv('CTDAPD Dataset.csv')

# Ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Handle missing values (only for numeric columns)
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Select 'Anomaly_Score' as the target variable
target_column = 'Anomaly_Score'
data = df[target_column]

# Normalize the data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data.values.reshape(-1, 1))

# Prepare the data for LSTM model
def create_dataset(dataset, time_step=1):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

# Define time step (lookback period for LSTM)
time_step = 30  # You can experiment with different time steps

# Create dataset
X, y = create_dataset(data_scaled, time_step)

# Reshape the input data to be compatible with LSTM input
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split the data into training and testing sets (80% training, 20% testing)
train_size = int(len(X) * 0.80)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model with more complexity and tuned parameters
from tensorflow.keras.optimizers import Adam

# Build the LSTM model with adjusted hyperparameters
model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(time_step, 1)))  # Increased LSTM units
model.add(LSTM(units=50, return_sequences=False))  # Added another LSTM layer
model.add(Dense(units=1))

# Compile model with a custom learning rate
optimizer = Adam(learning_rate=0.001)  # You can experiment with this value
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Train the model and keep track of training and validation loss
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))  # Increased epochs

# Plot training & validation loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Predict the future values (for next 1 year, if your data is daily)
n_days = 365  # Predict for the next 1 year
predicted_values = []

# Start prediction from the last available point
last_data = X[-1]

for i in range(n_days):
    predicted_value = model.predict(last_data.reshape(1, time_step, 1))
    predicted_values.append(predicted_value[0][0])
    last_data = np.append(last_data[1:], predicted_value, axis=0)

# Invert the scaling to get the actual values (back to original scale)
predicted_values_rescaled = scaler.inverse_transform(np.array(predicted_values).reshape(-1, 1))

# Plot the predictions for the next 1 year
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(data)), data, label="Actual Anomaly Scores")
plt.plot(np.arange(len(data), len(data) + n_days), predicted_values_rescaled, label="Predicted Anomaly Scores (Next Year)", color='red')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Anomaly Score')
plt.title('Anomaly Score Prediction for Next 1 Year')
plt.show()

# Output the predicted anomaly scores for the next 365 days
print("Predicted Anomaly Scores for Next 365 Days:")
print(predicted_values_rescaled)

# Evaluate the model performance (MAE and RMSE)
y_pred = model.predict(X_test)

# Inverse scaling to get the actual values
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate MAE and RMSE
mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
rmse = math.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))

print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")

# Plot Actual vs Predicted values
plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(y_test_rescaled):], y_test_rescaled, label='Actual Anomaly Scores')
plt.plot(df.index[-len(y_pred_rescaled):], y_pred_rescaled, label='Predicted Anomaly Scores', color='red')
plt.xlabel('Date')
plt.ylabel('Anomaly Score')
plt.title('Actual vs Predicted Anomaly Scores')
plt.legend()
plt.show()

# Plot residuals (difference between actual and predicted values)
residuals = y_test_rescaled.flatten() - y_pred_rescaled.flatten()

plt.figure(figsize=(12, 6))
plt.plot(df.index[-len(residuals):], residuals, label='Residuals', color='purple')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.title('Residuals between Actual and Predicted Anomaly Scores')
plt.legend()
plt.show()
import joblib

# After training the model
model_filename = 'lstm_model.joblib'

# Save the model using joblib
joblib.dump(model, model_filename)

# Save the scaler as well
scaler_filename = 's.joblib'
joblib.dump(scaler, scaler_filename)
```
